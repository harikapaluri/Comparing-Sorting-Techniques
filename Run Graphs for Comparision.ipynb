```python
import pandas as pd
df=pd.read_csv('ExecutionTimes.csv')
```

```python
df=df.groupby(['Sorting','Input Size'], as_index=False).mean()
```

```python
df.to_csv('new_data.csv')
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import interpolate

for name in df.Sorting.unique():
    
    temp_df = df.loc[df['Sorting'] == name]
    x=temp_df['Input Size']
    y=temp_df['Time in ms']

    x_new = np.linspace(1000, 50000, 300)

    a_BSpline = interpolate.make_interp_spline(x, y)

    y_new = a_BSpline(x_new)
    fig, ax = plt.subplots()
    ax.plot(x_new, y_new)
    ax.set_title(name)
    ax.set(xlabel='Input Size', ylabel='Time in ms')
    plt.show();
```
